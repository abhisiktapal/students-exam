```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
student_data=pd.read_csv('StudentsPerformance.csv')
print(student_data)
student_data.head(5)
```

```python
top=student_data['math score'].head(10)
pdf=pd.DataFrame(top)
ax=pdf.plot.bar()
ax.set_xlabel("student")
ax.set_ylabel("math score")
ax.set_title("Graph of math score of students")
```
